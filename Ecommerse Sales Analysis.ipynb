```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
d = pd.read_excel("./Superstore.xlsx")
print(d)
```

```python
d.isnull().sum()
```

```python
d.sort_values(['Profit','Sales'],axis=0,ascending=[False,False],inplace=True)
```

```python
fig = px.bar(d.head(100),x='City', y='Sales',color='Category',width=700,title='maximum and minimum sales in cities')
fig.show()
```

```python
fig = px.bar(d.head(20),x='Customer Name',y='Quantity'
             ,color='Profit',
             hover_data=['Discount','Product Name','Sub-Category'],
             width=1000,title='Product with quantity purchased by Customer')
fig.show()
```

```python
fig =px.pie(d,names='Sub-Category',values='Profit',title='Sub-Categories with profit using pie chart')
fig.show()
```
